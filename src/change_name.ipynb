```python
import glob
import os


def rename(f_path, new_name):
    filelist = glob.glob(f_path + "*.jpg")
    count = 0
    for file in filelist:
        print("File Count : ", count)
        filename = os.path.split(file)
        print(filename)
        new_filename = f_path + new_name + str(count + 1) + ".jpg"
        os.rename(f_path+filename[1], new_filename)
        print(new_filename)
        count = count + 1
```

```python
new_name = 'ww'
f_path = '../input/other/white/women/'

rename(f_path, new_name)
```
